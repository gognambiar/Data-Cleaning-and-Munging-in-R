# Activity 1
### In this activity xml is transformed into dataframe and then to csv

```r
library(XML)
```

```r
xml.url <- "http://aiweb.cs.washington.edu/research/projects/xmltk/xmldata/data/courses/reed.xml"
```

```r
z = xmlToDataFrame(xml.url)
doc = xmlParse(xml.url)
s = xmlToList(doc)
```

```r
stt = c()
ett = c()
plb = c()
plr = c()
for(i in 1:length(s))
{
if(is.null(s[i]$course$time$start_time)&&is.null(s[i]$course$time$end_time))
{
    stt <- c(stt,"NA")
    ett <- c(ett,"NA")
}
    
else if(is.null(s[i]$course$time$start_time))
{
    stt <- c(stt,"NA")
    ett <- c(ett,s[i]$course$time$end_time)
}
else if(is.null(s[i]$course$time$end_time))
{
    stt <- c(stt,s[i]$course$time$start_time)
    ett <- c(ett,"NA")
}
else
{
stt <- c(stt,s[i]$course$time$start_time)
ett <- c(ett,s[i]$course$time$end_time)
}

if(is.null(s[i]$course$place$building)&&is.null(s[i]$course$place$room))
{
    plb <- c(plb,"NA")
    plr <- c(plr,"NA")
}

else if(is.null(s[i]$course$place$building))
{
    plb <- c(plb,"NA")
    plr <- c(plr,s[i]$course$place$room)
}
else if(is.null(s[i]$course$place$room))
{
    plb <- c(plb,s[i]$course$place$building)
    plr <- c(plr,"NA")
}
    
else
{
plb <- c(plb,s[i]$course$place$building)
plr <- c(plr,s[i]$course$place$room)
}
    
}
```

```r
z$start_time <- stt
z$end_time <- ett
z$place_building <- plb
z$place_room <- plr
```

```r
drop <- c("time","place")
z = z[,!(names(z) %in% drop)]
```

```r
write.csv(file="str1.csv", x=z)
```
